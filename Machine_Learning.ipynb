```python
# Image classification with a pretrained MobileNet model (Python)

import tensorflow as tf
from tensorflow.keras.applications import MobileNet
from tensorflow.keras.applications.mobilenet import preprocess_input, decode_predictions
from tensorflow.keras.preprocessing import image
import numpy as np
import matplotlib.pyplot as plt

# Load pretrained MobileNet model
model = MobileNet(weights="imagenet")

# Load and display image (upload pic1.jpg to Colab)
img_path = "pic1.jpg"
img = image.load_img(img_path, target_size=(224, 224))

plt.imshow(img)
plt.axis("off")

# Prepare image for prediction
img_array = image.img_to_array(img)
img_array = np.expand_dims(img_array, axis=0)
img_array = preprocess_input(img_array)

# Classify image
predictions = model.predict(img_array)
decoded_predictions = decode_predictions(predictions, top=5)[0]

for _, label, confidence in decoded_predictions:
    print(f"{label}: {confidence:.2%}")
```

## Perceptrons

A **Perceptron** is an artificial neuron.

It is the simplest possible neural network and one of the foundational building blocks of Machine Learning and Artificial Intelligence.

Perceptrons were introduced in 1957 by **Frank Rosenblatt**, inspired by how biological neurons receive inputs, process them, and produce decisions.


## Frank Rosenblatt

Frank Rosenblatt (1928â€“1971) was an American psychologist and AI researcher.

In 1957, he created the first perceptron program on an IBM 704 computer.

His idea was that machines could learn by adjusting the importance (weights) of inputs, similar to how the human brain learns from experience.


## The Perceptron Concept

A perceptron:
- Takes multiple binary inputs (0 or 1)
- Multiplies each input by a weight
- Sums the weighted values
- Compares the sum to a threshold
- Outputs either 0 or 1 (No / Yes)

Mathematically:

y = 1 if Î£(x Â· w) > threshold  
y = 0 otherwise


## Example: Should You Go to a Concert?

Imagine a perceptron deciding whether you should go to a concert.

Inputs represent facts:
- Is the artist good?
- Is the weather good?
- Will a friend come?
- Is food served?
- Is alcohol served?

Each input has a weight that represents its importance.


### Inputs and Weights

| Criteria            | Input | Weight |
|---------------------|-------|--------|
| Artist is good      | 1     | 0.7    |
| Weather is good     | 0     | 0.6    |
| Friend will come    | 1     | 0.5    |
| Food is served      | 0     | 0.3    |
| Alcohol is served   | 1     | 0.4    |

Threshold = **1.5**


```python
# Threshold
threshold = 1.5

# Inputs (binary)
inputs = [1, 0, 1, 0, 1]

# Corresponding weights
weights = [0.7, 0.6, 0.5, 0.3, 0.4]

# Weighted sum
weighted_sum = 0
for x, w in zip(inputs, weights):
    weighted_sum += x * w

# Activation function
output = 1 if weighted_sum > threshold else 0

weighted_sum, output
```

## How the Decision Is Made

1. Each input is multiplied by its weight
2. All results are summed
3. The sum is compared to the threshold
4. If the sum exceeds the threshold, the perceptron outputs 1
5. Otherwise, it outputs 0

In this example:
- Weighted sum = 1.6
- Threshold = 1.5
- Output = 1 (Yes, go to the concert)


## Perceptron Components

A perceptron consists of:

2. Weights

## Linear Binary Classification

Perceptrons are **linear binary classifiers**.

They separate data using a straight line (in 2D) or a hyperplane (in higher dimensions).

Points are classified as:
- Above the line â†’ Class 1
- Below the line â†’ Class 0


## Idea Behind Pattern Classification

Imagine a straight line drawn through a scatter plot of points.

Even if we do not know the equation of the line, a perceptron can learn to classify points as:
- Above the line
- Below the line


```python
import numpy as np
import matplotlib.pyplot as plt

# Number of points
num_points = 500

# Generate random x and y values
x_points = np.random.rand(num_points) * 100
y_points = np.random.rand(num_points) * 100
```

## Creating a Line Function

We define a linear function:

y = 1.2x + 50

This line represents the decision boundary.


```python
def f(x):
    return 1.2 * x + 50
```

## Computing the Correct Answers

For each point:
- Output = 1 if the point is above the line
- Output = 0 if the point is below the line

These are the **true labels** used for training.


```python
desired = []

for x, y in zip(x_points, y_points):
    if y > f(x):
        desired.append(1)
    else:
        desired.append(0)

desired = np.array(desired)
```

## Visualizing the Data

We plot:
- Blue points â†’ Below the line
- Black points â†’ Above the line
- Red line â†’ Decision boundary


```python
plt.figure(figsize=(8, 6))

# Plot points
for i in range(num_points):
    if desired[i] == 1:
        plt.scatter(x_points[i], y_points[i], color="black", s=10)
    else:
        plt.scatter(x_points[i], y_points[i], color="blue", s=10)

# Plot decision line
x_line = np.linspace(0, 100, 100)
y_line = f(x_line)
plt.plot(x_line, y_line, color="red", linewidth=2)

plt.xlabel("X Value")
plt.ylabel("Y Value")
plt.title("Pattern Recognition with a Linear Boundary")
plt.show()
```

## What We Achieved

- Generated random data points
- Defined a linear decision boundary
- Classified points as above or below the line
- Visualized the classification

This is exactly the kind of problem a perceptron is designed to solve.


## Next Step: Training the Perceptron

So far, we used the correct answers directly.

Next, we will:
- Initialize random weights
- Let the perceptron make predictions
- Measure errors
- Update weights using a learning rule

This is called **Perceptron Training**.


## Training a Perceptron

Training a perceptron means adjusting its weights so it can correctly classify data.

The perceptron learns by:
- Making a prediction
- Comparing it to the correct answer
- Adjusting weights when it is wrong


## Training Task

Imagine a straight line separating points in a 2D space.

The goal is to train a perceptron to:
- Output `1` for points above the line
- Output `0` for points below the line


## Creating a Perceptron Object

The perceptron needs:
- A number of inputs
- A learning rate
- Random initial weights
- A bias term


```python
import numpy as np
```

```python
class Perceptron:
    def __init__(self, num_inputs, learning_rate=0.00001):
        self.learning_rate = learning_rate
        self.bias = 1
        # +1 for bias weight
        self.weights = np.random.uniform(-1, 1, num_inputs + 1)

    def activate(self, inputs):
        summation = np.dot(inputs, self.weights)
        return 1 if summation > 0 else 0

    def train(self, inputs, desired):
        guess = self.activate(inputs)
        error = desired - guess

        if error != 0:
            self.weights += self.learning_rate * error * inputs
```

## The Learning Rate

The learning rate controls how much the weights change after a mistake.

- Too large â†’ unstable learning
- Too small â†’ very slow learning


## The Bias

The bias allows the perceptron to shift the decision boundary.

It is implemented as:
- An extra input with a constant value of `1`


## Generating Training Data

We generate random (x, y) points and define a linear function:

y = 1.2x + 50


```python
import matplotlib.pyplot as plt

num_points = 500

x_points = np.random.rand(num_points) * 100
y_points = np.random.rand(num_points) * 100

def f(x):
    return 1.2 * x + 50
```

## Computing the Correct Answers

If a point is above the line â†’ label = 1  
If a point is below the line â†’ label = 0


```python
desired = np.array([
    1 if y > f(x) else 0
    for x, y in zip(x_points, y_points)
])
```

## Creating the Perceptron

We use:
- 2 inputs (x and y)
- A small learning rate


```python
ptron = Perceptron(num_inputs=2, learning_rate=0.00001)
```

## Training the Perceptron

The perceptron is trained over many iterations.

Each iteration adjusts the weights to reduce error.


```python
epochs = 10000

for _ in range(epochs):
    for x, y, label in zip(x_points, y_points, desired):
        inputs = np.array([x, y, ptron.bias])
        ptron.train(inputs, label)
```

## Visualizing the Result

After training:
- Correctly classified points appear separated
- The perceptron has learned the pattern


```python
plt.figure(figsize=(8, 6))

for x, y in zip(x_points, y_points):
    guess = ptron.activate(np.array([x, y, ptron.bias]))
    color = "black" if guess == 1 else "blue"
    plt.scatter(x, y, color=color, s=10)

# Plot the true line
x_line = np.linspace(0, 100, 100)
plt.plot(x_line, f(x_line), color="red", linewidth=2)

plt.xlabel("X Value")
plt.ylabel("Y Value")
plt.title("Trained Perceptron Classification")
plt.show()
```

## Visual Testing

```python
plt.figure(figsize=(8, 6))

for x, y in zip(x_test, y_test):
    guess = ptron.activate(np.array([x, y, ptron.bias]))
    color = "black" if guess == 1 else "blue"
    plt.scatter(x, y, color=color, s=10)

# True separating line
x_line = np.linspace(0, 100, 100)
plt.plot(x_line, f(x_line), color="red", linewidth=2)

plt.xlabel("X Value")
plt.ylabel("Y Value")
plt.title("Perceptron Predictions on Unseen Data")
plt.show()
```

## Counting Errors

To evaluate performance, we count how many predictions are incorrect.


```python
errors = 0

for x, y in zip(x_test, y_test):
    guess = ptron.activate(np.array([x, y, ptron.bias]))
    actual = 1 if y > f(x) else 0
    if guess != actual:
        errors += 1
```

## Error Results


```python
error_rate = errors / test_points

print(f"Total test points: {test_points}")
print(f"Number of errors: {errors}")
print(f"Error rate: {error_rate:.4f}")
```

## Interpreting the Results

- A low error rate means the perceptron learned the pattern well
- A high error rate means the perceptron needs improvement


## Tuning the Perceptron

If performance is poor, try:
- Adjusting the learning rate
- Increasing the number of training points
- Increasing the number of training iterations (epochs)


## Key Takeaway

Training teaches the perceptron.

Testing proves whether it actually learned.


# Machine Learning: Learning is Looping

Machine learning models learn by looping over data multiple times.

During each iteration:
- The model updates its weights
- The goal is to reduce error (cost)

Training stops when further iterations no longer reduce the cost.


## Gradient Descent

Gradient Descent is a common optimization algorithm.

It is used here to train a **Linear Regression model** of the form:

y = wx + b

The algorithm adjusts:
- w â†’ weight (slope)
- b â†’ bias (intercept)

until the best-fitting line is found.


## Trainer Object

We create a Trainer object that:
- Accepts x and y data
- Uses gradient descent
- Minimizes prediction error


```python
import numpy as np
```

## Creating the Trainer Class

Initial values:
- Weight = 0
- Bias = 1
- Learning rate = 0.00001


```python
class Trainer:
    def __init__(self, x_array, y_array):
        self.x = np.array(x_array)
        self.y = np.array(y_array)
        self.points = len(self.x)
        self.learnc = 0.00001
        self.weight = 0.0
        self.bias = 1.0
        self.cost = None
```

## Cost Function (Error Function)

The cost function measures how well the model fits the data.

It computes the **mean squared error** between:
- Actual y values
- Predicted y values (wx + b)


```python
    def cost_error(self):
        predictions = self.weight * self.x + self.bias
        errors = self.y - predictions
        return np.mean(errors ** 2)
```

## Training Function

Training runs gradient descent for a fixed number of iterations.

After training, the final cost is stored.


```python
    def train(self, iterations):
        for _ in range(iterations):
            self.update_weights()
        self.cost = self.cost_error()
```

## Updating Weights and Bias

The gradients are computed using partial derivatives:

- Weight gradient
- Bias gradient

These gradients determine the direction of improvement.


```python
    def update_weights(self):
        predictions = self.weight * self.x + self.bias
        errors = self.y - predictions

        w_deriv = -2 * np.mean(errors * self.x)
        b_deriv = -2 * np.mean(errors)

        self.weight -= self.learnc * w_deriv
        self.bias -= self.learnc * b_deriv
```

## Full Trainer Class Summary

The Trainer:
- Uses gradient descent
- Learns the best-fit line
- Minimizes mean squared error


## Key Takeaways

- Learning happens through looping
- Gradient descent reduces error step-by-step
- Linear regression can be learned without formulas

# Linear Regression

A regression is a method used to determine the relationship between:
- an output variable (**y**)
- one or more input variables (**x**)

In **Machine Learning**, Linear Regression is a **supervised learning algorithm**.


## Scatter Plot

Before fitting a model, we visualize the relationship between house size
and price using a scatter plot.


```python
import matplotlib.pyplot as plt

# Data
x = [50, 60, 70, 80, 90, 100, 110, 120, 130, 140, 150]
y = [7, 8, 8, 9, 9, 9, 10, 11, 14, 14, 15]

# Scatter plot
plt.scatter(x, y)
plt.xlabel("Square Meters")
plt.ylabel("Price in Millions")
plt.title("House Prices vs Size")
plt.show()
```

## Predicting Values

From the scattered data, we want to predict future prices.

This can be done by:
- Drawing a linear graph
- Modeling a linear relationship
- Using linear regression


## Linear Graphs

A linear graph can be written as:

\[
y = ax + b
\]

Where:
- **y** is the predicted value
- **x** is the input value
- **a** is the slope
- **b** is the intercept


## Simple Linear Model (Slope Only)

First, we calculate an average slope assuming the intercept is zero.


```python
# Calculate simple slope
x_sum = sum(x)
y_sum = sum(y)
slope_simple = y_sum / x_sum

# Generate prediction line
x_vals = list(range(50, 151))
y_vals = [slope_simple * xi for xi in x_vals]

# Plot
plt.scatter(x, y)
plt.plot(x_vals, y_vals)
plt.xlabel("Square Meters")
plt.ylabel("Price in Millions")
plt.title("Simple Linear Model")
plt.show()
```

In this model:
- The slope is calculated as an average
- The intercept is assumed to be **0**
- This is a very rough approximation


## Linear Regression Function

Now we calculate the **best-fit line** using the linear regression formula.


```python
# Number of data points
n = len(x)

# Calculate sums
x_sum = sum(x)
y_sum = sum(y)
xx_sum = sum([xi**2 for xi in x])
xy_sum = sum([x[i] * y[i] for i in range(n)])

# Calculate slope and intercept
slope = (n * xy_sum - x_sum * y_sum) / (n * xx_sum - x_sum**2)
intercept = (y_sum / n) - (slope * x_sum) / n

slope, intercept
```

Using the calculated slope and intercept, we can generate predictions.


```python
# Generate regression line
x_vals = list(range(50, 151))
y_vals = [slope * xi + intercept for xi in x_vals]

# Plot
plt.scatter(x, y)
plt.plot(x_vals, y_vals)
plt.xlabel("Square Meters")
plt.ylabel("Price in Millions")
plt.title("Linear Regression Model")
plt.show()
```

## Interpretation

- The regression line minimizes overall error
- This model fits the data better than the simple slope model
- Linear regression assumes a linear relationship


## Polynomial Regression

If the data does not follow a straight line, a **polynomial regression**
may be more appropriate.

Polynomial regression allows curved relationships between x and y.
